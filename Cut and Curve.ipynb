```python
#correct code
import pandas as pd

input_file = input('Please enter the name of the input Excel file with extension (Ex-abc.xlsx): ')

# Accept inputs for curve data from User
curve_sheet_name = input('Please enter the name of the worksheet containing Curve data (Ex-Sheet4): ')
curve_start_col = input('Please enter the column containing Starting Chainage of Curve data (Ex-B): ')
curve_end_col = input('Please enter the column containing Ending Chainage of Curve data (Ex-C): ')

# Accept inputs for cutting data from User
cutting_sheet_name = input('Please enter the name of the worksheet containing Cutting data (Ex-Sheet1): ')
cutting_start_col = input('Please enter the column containing Starting Chainage of Cutting data (A): ')
cutting_end_col = input('Please enter the column containing Ending Chainage of Cutting data (B): ')

# Read Excel data into DataFrames
curve_data = pd.read_excel(input_file, sheet_name=curve_sheet_name)
cutting_data = pd.read_excel(input_file, sheet_name=cutting_sheet_name)

# Create 2-D lists cuts and curves
cuts = cutting_data[[cutting_start_col, cutting_end_col]].values.tolist()
curves = curve_data[[curve_start_col, curve_end_col]].values.tolist()

# Print the resulting lists (for verification)
print("Cuts:", cuts)
print("Curves:", curves)
intersections = []
i = j = 0  # Initialize pointers for both lists.
# Continue while neither list is fully traversed.
while i < len(cuts) and j < len(curves):
    # Find the furthest start time and earliest end time to conveniently check 
    # if the intervals intersect.
    maxStart = max(cuts[i][0], curves[j][0])
    minEnd = min(cuts[i][1], curves[j][1])
    # If the intervals intersect, add the intersection to the output.
    if maxStart <= minEnd:
        intersections.append([maxStart, minEnd])
    # Move the pointer forward for the interval that ends first.
    if cuts[i][1] < curves[j][1]:
        i += 1
    else:
        j += 1
print("Intersections:")
print(intersections)
```

```python

degree_col = input('Please enter the name of the column containing degrees in Curve data: ')

# Fetch the degrees for each intersection
for intersection in intersections:
    # Identify the row in the 'curve_data' DataFrame where the intersection lies
    intersecting_rows = curve_data[
        (curve_data[curve_start_col] <= intersection[0]) & (curve_data[curve_end_col] >= intersection[1])
    ]

    # Check if there are any intersecting rows
    if not intersecting_rows.empty:
        # Print the degree value from the identified row
        degree_value = intersecting_rows.iloc[0][degree_col]
        print(f"Degree for Interval {intersection[0]} - {intersection[1]}: {degree_value}")
    else:
        print(f"No matching row found for Interval {intersection[0]} - {intersection[1]}")
```
